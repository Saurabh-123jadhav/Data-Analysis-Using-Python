```python

df=pd.read_csv(r"C:\Users\Vishu\Downloads\Python project\Python_Diwali_Sales_Analysis\Diwali Sales Data.csv", encoding="unicode_escape")
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

#From above graph we can say that most of the buyers are female And even the purchasing power of women is greater than men

Age Analsis

```python
sales_age_count=sns.countplot(x='Age Group', data=df, hue='Gender')
for bar in sales_age_count.containers:
    sales_age_count.bar_label(bar)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group', palette='viridis', legend=False)
```

#From the above graph we can see that most of the buyers are of age group of 26-35 yrs and female

Sates analysis


```python
sales_data=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25, 5)})
sns.barplot(x='State', y='Orders', data=sales_data, hue='State', palette='viridis', legend=False)
```

```python
states_amount=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State', y='Amount', data=states_amount, hue='State', palette='viridis')
```

#From the above graphs we can see that the most of the orders and sales/amount are from Uttar Pradesh, Maharashtra, Karnataka

```python
df.columns
```

Marital status Analysis

```python
sns.set(rc={'figure.figsize':(5,5)})
ax=sns.countplot(x='Marital_Status', data=df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

#From the above graphs we can say that the most of the buyers are married(women) and they have most purchasing power

OCCUPATION ANALYSIS

```python
occupation_order = df['Occupation'].value_counts().index
ax=sns.countplot(x='Occupation', data=df, hue='Occupation', palette='viridis', legend=False, order=occupation_order)
sns.set(rc={'figure.figsize':(27,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_occupation=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(27,5)})
sns.barplot(x='Occupation', y='Amount', data=sales_occupation, hue='Occupation', palette='viridis', legend=False)
```

#From the above graph most of the buyers are working in IT , Healthcare anf Aviation sector

PRODUCT CATEGORY ANALYSIS

```python
top_10 = df['Product_Category'].value_counts(ascending=False).head(10)

# Plotting the top 10 categories
sns.set(rc={'figure.figsize': (27, 5)})
ax = sns.barplot(x=top_10.index, y=top_10.values)

# Adding bar labels
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
sales_product=df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='Product_Category', y='Orders', data=sales_product, hue='Product_Category', palette='viridis', legend=False)
```

```python
sales_product=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='Product_Category', y='Amount', data=sales_product, hue='Product_Category', palette='viridis', legend=False)
```


#From the above graphs, we can conclude that most orders are for clothing, followed by food and electronics. However, food products generated higher sales than clothing items.

```python
p_idsales=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='Product_ID', y='Orders', data=p_idsales, hue='Product_ID', palette='viridis', legend=False)
```
