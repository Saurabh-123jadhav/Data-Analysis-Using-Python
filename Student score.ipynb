# GENDER DISTRIBUTION

```python
sns.set(rc={'figure.figsize':(5,6)})
plt.figure(figsize=(5,5))
plt.title("Gender Distribution", {'fontsize':18, "fontweight":"bold"})
ax=sns.countplot(x="Gender", data=df, hue="Gender", palette="viridis", legend=False)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
gb=df.groupby('ParentEduc').agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
print(gb)
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(gb, annot=True)
plt.title("Relationship between Parents Eduction and childrens score", {'fontsize':18, "fontweight":"bold"})
```

#From the above graph we can say that higher the education of parents higher the kids performance and
lower the parents education lower the kids performance.

# DISTRIBUTION OF ETHINIC GROUP

```python
groupA=df.loc[(df['EthnicGroup']== 'group A')].count()
groupB=df.loc[(df['EthnicGroup']=='group B')].count()
groupC=df.loc[(df['EthnicGroup']=='group C')].count()
groupD=df.loc[(df['EthnicGroup']=='group D')].count()
groupE=df.loc[(df['EthnicGroup']=='group E')].count()


    
l=['group A', 'groupB', 'groupC', 'groupD', 'groupE']
mylist=[groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
plt.pie(mylist, labels=l, autopct='%1.2f%%', shadow=True, startangle=90, explode=(0,0,0.1,0,0))
plt.title("DISTRIBUTION OF ETHNIC GROUP")
plt.show()
```

```python
ax=sns.countplot(data=df, x='EthnicGroup').bar_label(ax.containers[0])
plt.title("COUNT OF ETHNIC GROUP", fontdict={'fontweight':'bold', 'fontsize':18}) 
plt.show()
```
